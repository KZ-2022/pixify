```python
reader = easyocr.Reader(['en'])
result = reader.readtext(IMAGE_PATH,paragraph="True")
result
```

```python
IMAGE_PATH2 = 'mounjaro2.png'
reader = easyocr.Reader(['en'])
result = reader.readtext(IMAGE_PATH2,paragraph="False")
result
```

```python
IMAGE_PATH3 = 'Picture1.jpg'
IMAGE_PATH3 = 'Picture2.png'

reader = easyocr.Reader(['en','hi'])
result = reader.readtext(IMAGE_PATH3, paragraph="True", detail=0)
result
```

```python
result = "\n".join(result)
result
```
